```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
from time import time
```

```python
def dbscan(X, clusters, eps):
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(X)

    t0 = time()
    db = DBSCAN(eps=eps, min_samples=1)
    pred = db.fit_predict(X)

    print("Clustering finished for %d clusters." % clusters)
    print("Done in %f seconds." % (time() - t0))

    return pred
```

```python
df = pd.read_excel('../dataset/clean_dataset++.xlsx')
df
```

```python
t0 = time()
df['Cluster 1'] = dbscan(df['Text'], 3273, 0.513)
df['Cluster 2'] = dbscan(df['Text'], 1564, 0.968)
print("Total: %f seconds." % (time() - t0))
```

```python
df.to_excel("../dataset/db++.xlsx", index=False)
```
